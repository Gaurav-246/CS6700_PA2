```python
import numpy as np
import matplotlib.pyplot as plt
```

## CartPole-v1 REINFORCE

```python
n = np.load(r"C:\Users\kathi\Downloads\SEM-8\CS6700\PA2\final\all_scores_array_cp_wtd0b.npy")
m = np.load(r"C:\Users\kathi\Downloads\SEM-8\CS6700\PA2\final\all_scores_array_cp_wob.npy")
```

```python
for i in range(5):
    plt.plot(m[i])
```

```python
for i in range(5):
    plt.plot(n[i])
```

```python
m1 = np.mean(m, axis=0)
s1 = np.std(m, axis=0)

fig, ax = plt.subplots(1,1, figsize=(15,9))
plt.plot(m1, label='w/o Baseline')
ax.fill_between(range(m1.shape[0]), m1 - s1, m1 + s1, alpha=0.3 )
plt.xlabel(r'Episodes$\rightarrow$', fontsize=15)
plt.ylabel(r'Rewards$\rightarrow$', fontsize=15)
plt.tick_params(axis='x',labelsize=13)
plt.tick_params(axis='y',labelsize=13)
plt.title('CartPole-v1 - MC REINFORCE - Plot of Rewards per Episode', fontsize=15)
plt.legend(fontsize=15)
```

```python
s2 = np.std(n, axis=0)
m2 = np.mean(n, axis=0)
fig, ax = plt.subplots(1,1, figsize=(15,9))
plt.plot(m2, label='Baseline', color='orange')
ax.fill_between(range(m2.shape[0]), m2 - s2, m2 + s2, alpha=0.3, color='orange' )
plt.xlabel(r'Episodes$\rightarrow$', fontsize=15)
plt.ylabel(r'Rewards$\rightarrow$', fontsize=15)
plt.tick_params(axis='x',labelsize=13)
plt.tick_params(axis='y',labelsize=13)
plt.title('CartPole-v1 - MC REINFORCE - Plot of Rewards per Episode', fontsize=15)
plt.legend(fontsize=15)
```
