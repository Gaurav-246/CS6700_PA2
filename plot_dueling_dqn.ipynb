```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_rolling_averages(array, window_size=100):
    num_elements = len(array)
    rolling_averages = []
    for i in range(num_elements - window_size + 1):
        chunk = array[i:i+window_size]
        average = np.mean(chunk)
        rolling_averages.append(average)
    return rolling_averages
```

## CartPole-v1 - Dueling DQN Plots

```python
type1 = []
type2 = []
for i in range(5):
    type1.append(np.load(r"C:\Users\g\Downloads\CartPole_Type_1_Exp_"+str(i+1)+".npy"))
    type2.append(np.load(r"C:\Users\g\Downloads\CartPole_Type_2_Exp_"+str(i+1)+".npy"))
```

```python
x1 = np.mean(type1, axis=0)
x2 = np.mean(type2, axis=0)
y1 = np.std(type1, axis=0)
y2 = np.std(type2, axis=0)

fig, ax = plt.subplots(1,1, figsize=(15,9))
plt.plot(x1, label='Type 1')
ax.fill_between(range(x1.shape[0]), x1 - y1, x1 + y1, alpha=0.3 )
plt.plot(x2, label='Type 2')
ax.fill_between(range(x2.shape[0]), x2 - y2, x2 + y2, alpha=0.3 )
plt.xlabel(r'Episodes$\rightarrow$', fontsize=15)
plt.ylabel(r'Rewards$\rightarrow$', fontsize=15)
plt.tick_params(axis='x',labelsize=13)
plt.tick_params(axis='y',labelsize=13)
plt.title('CartPole-v1 - Dueling DQN - Plot of Rewards per Episode', fontsize=15)
plt.legend(fontsize=15)
```

```python
z1_rolling = []
z2_rolling = []

for item in type1:
    z1_rolling.append(calculate_rolling_averages(item))
for item in type2:
    z2_rolling.append(calculate_rolling_averages(item))

x1_avg_rolling = np.mean(z1_rolling, axis=0)
x2_avg_rolling = np.mean(z2_rolling, axis=0)
y1_avg_rolling = np.std(z1_rolling, axis=0)
y2_avg_rolling = np.std(z2_rolling, axis=0)
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,9))
plt.plot(x1_avg_rolling, label='Type 1')
ax.fill_between(range(x1_avg_rolling.shape[0]), x1_avg_rolling - y1_avg_rolling, x1_avg_rolling + y1_avg_rolling, alpha=0.3 )
plt.plot(x2_avg_rolling, label='Type 2')
ax.fill_between(range(x2_avg_rolling.shape[0]), x2_avg_rolling - y2_avg_rolling, x2_avg_rolling + y2_avg_rolling, alpha=0.3 )
plt.plot(range(x1_avg_rolling.shape[0]), [195]*x1_avg_rolling.shape[0], linestyle='--', color='red', label='195')
plt.xlabel(r'Episodes$\rightarrow$',fontsize=15)
plt.ylabel(r'Rewards$\rightarrow$',fontsize=15)
plt.tick_params(axis='x',labelsize=13)
plt.tick_params(axis='y',labelsize=13)
plt.title('CartPole-v1 - Dueling DQN  - Moving Avg. Reward per 100 episodes',fontsize=15)
plt.legend(fontsize=15)
```
